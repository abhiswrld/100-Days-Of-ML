# Types of feature engineering

### The process of taking in raw data, and creating new information that is useful for predicting the outcome 

#### 1. Feature Transformation
* Missing Value Imputation
* Handling Categorical Features
* Outlier Detection
* Feature Scaling

#### 2. Feature Construction
* We want to manually create new features in order to better train our model

#### 3. Feature Selection
* Pick only features that have something to do with the final output

#### 4. Feature Extraction
* Extract new features progamatically

# Missing Values

* If there are a lot of values missing, then we must fill it. One way to do so is to average the column and apply that value to missing cells
- For categorical data we can replace with most frequent value
* If only a few cells are missing values and the dataset is big enough, we can ignore them/remove them from the dataset

# Categorical Values

* We have to transform categorical values to numerical values for sci-kit learn to accept them
* For example: we can use one-hot encoding
