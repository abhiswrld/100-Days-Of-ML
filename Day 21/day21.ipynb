# Multivariate/Bivariate Analysis

```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
titanic = pd.read_csv('train.csv')
```

```python
titanic.head()
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

### Use a scatterplot to compare (bivariate) variables

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker')
```

```python
sns.barplot(data=titanic, x='Pclass', y='Age')
```

```python
sns.displot(titanic[titanic['Survived']==0]['Age'], kde=True)
sns.displot(titanic[titanic['Survived']==1]['Age'], kde=True)
```

```python
titanic.head()
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean()*100
```

```python
sns.clustermap(pd.crosstab(titanic['SibSp'], titanic['Survived']))
```

```python
iris.head()
```

```python
sns.pairplot(iris, hue='species')
```

```python
flights.head()
```

```python
sns.lineplot(data=flights, x='year', y='passengers', hue='month')
```

```python
new = flights.groupby(['year', 'month'])['passengers'].sum().reset_index()
```

```python
sns.lineplot(data=new, x='year', y='passengers', hue='month')
```

```python
sns.heatmap(flights.pivot_table(values='passengers', index='month', columns='year'))
```
