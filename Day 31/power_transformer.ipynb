# Power Transformer

#### Box-Cox - Convert any distribution to a normal distribution
#### Yeo-Johnson

#### For Box-Cox, we need to calculate lambda, which varies over the range of -5 to 5, and in the process of finding the value we iterate through every single possible value. Finally we choose the most optimal value. n has to be > 0

#### For Yeo-Johnson, we can do everything that Box-Cox does, but we can also apply it to negative numbers.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn.preprocessing import PowerTransformer

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('Concrete Data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
X = df.drop(columns=['Strength'])
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Apply linear regression without transformation

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
r2_score(y_test, y_pred)
```

```python
lr = LinearRegression()
np.mean(cross_val_score(lr, X, y, scoring="r2"))
```

```python
# Applying Box-Cox transformation

pt = PowerTransformer('box-cox')
X_train_transformed = pt.fit_transform(X_train+0.00001)
X_test_transformed = pt.transform(X_test+0.00001)

pd.DataFrame({'cols':X_train.columns,'box_cox_lambdas':pt.lambdas_})
```

```python
lr = LinearRegression()
lr.fit(X_train_transformed, y_train)

y_pred2 = lr.predict(X_test_transformed)

r2_score(y_test, y_pred2)
```

```python
pt = PowerTransformer('box-cox')
X_transformed = pt.fit_transform(X+0.00001)
lr = LinearRegression()
np.mean(cross_val_score(lr, X_transformed, y, scoring="r2"))
```

# Yeo-Johnson

```python
pt1 = PowerTransformer('yeo-johnson')
X_train_transformed2 = pt1.fit_transform(X_train)
X_test_transformed2 = pt1.transform(X_test)

lr = LinearRegression()
lr.fit(X_train_transformed2, y_train)

y_pred3 = lr.predict(X_test_transformed2)

print(r2_score(y_test, y_pred3))
```
