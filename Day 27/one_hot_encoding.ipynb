#### Handle nominal categorical data using OHE

#### Anytime we have categorical data for variables, we must convert them into numbers so that the ML algorithms can understand the variables.
#### We use OHE for nominal categorical variables 

<p> The main way to convert the strings to numerical's is basically changing them into vectors
<p> We can do this by creating new columns for each category that we have, and then for each value we can keep it 1, and then the rest of the variables will be 0 in that row.

```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df = pd.read_csv('cars.csv')
```

```python
df.head()
```

```python
df['brand'].value_counts()
```

# OHE Using Pandas

```python
pd.get_dummies(df, columns=['fuel', 'owner'])
```

# K-1 OHE

```python
pd.get_dummies(df, columns=['fuel', 'owner'], drop_first=True)
```

# OHE Using Sklearn

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:4], df.iloc[:,-1], test_size=0.2, random_state=0)
```

# OHE with Top Categories

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python
df['brand_grouped'] = df['brand'].replace(repl, 'uncommon')
pd.get_dummies(df['brand_grouped'])
```
