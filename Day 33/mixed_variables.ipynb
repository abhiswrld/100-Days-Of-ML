# Mixed Data

#### What is it? If the data that is in a dataset, there is both numerical + categorical data in the same column.
#### Titanic - "Cabin" column has values such as "B5", "C24", "D41" etc, so there is both categorical and numerical
#### What we can do now is just split the character and the number into 2 columns.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Titanic Data.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title("Passengers travelling with _ amount of people")
```

```python
#extract numerical part

df['number_numerical'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
```

```python
#extract categorical part

df['number_categorical'] = np.where(df['number_numerical'].isnull(), df['number'], np.nan)

df.head()
```

```python
df['Cabin'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # regular expression for numerical part
df['cabin_cat'] = df['Cabin'].str[0] #first letter

df.head()
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
