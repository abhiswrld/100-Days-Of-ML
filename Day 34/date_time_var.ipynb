# Working with dates

```python
# converting to datetime object type
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

## 1. Extract year

```python
date['date_year'] = date['date'].dt.year

date.head()
```

## 2. Extract month

```python
date['date_month'] = date['date'].dt.month

date.head()
```

## 3. Extract Day

```python
date['date_day'] = date['date'].dt.day

date.head()
```

## 4. Day of week

```python
# Monday is 0th index

date['date_dow'] = date['date'].dt.day_name()

date.head()
```

## 5. Day is a weekend?

```python
date['date_is_weekend'] = np.where(date['date_dow'].isin(['Saturday', 'Sunday']), 1,0)

date.head()
```

## 6. Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week

date.head()
```

## 7. Extract time elapsed between dates

```python
import datetime

today = datetime.datetime.today()

today
```
