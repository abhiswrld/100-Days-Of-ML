```python
df = pd.DataFrame()
```

```python
import time

for i in range(1, 550):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=6b4bf614172e96c377e0fc2b299dc5eb&language=en-US&page=' + str(i))
    if 'results' not in response.json():
        print(f"Error on page {i}: {response.json()}")
        break

    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'vote_average', 'vote_count', 'release_date', 'original_language']]
    df = pd.concat([df, temp_df], ignore_index=True)

    time.sleep(1)
```

```python
# Check for duplicates
df.duplicated(subset=['id']).sum()

# If there are duplicates, remove them
df = df.drop_duplicates(subset=['id'], keep='first')

# Check shape again
df.shape
```
