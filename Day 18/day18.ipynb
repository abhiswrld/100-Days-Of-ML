# Web Scraping


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
webpage = requests.get('https://quotes.toscrape.com/page/1/').text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
print(soup.prettify())
```

```python
soup.find_all('h1')[0].text
```

```python
for i in soup.find_all('span', class_='text'):
    print(i.text)
```

```python
for i in soup.find_all('small', class_='author'):
    print(i.text)
```

```python
for i in soup.find_all('a', class_='tag'):
    print(i.text)
```

```python
df = pd.DataFrame(columns=['Quote', 'Author', 'Tags'])

quotes = soup.find_all('div', class_='quote')

for quote in quotes:
    text = quote.find('span', class_='text').text
    author = quote.find('small', class_='author').text
    tags = [tag.text for tag in quote.find_all('a', class_='tag')]

    df = pd.concat([df, pd.DataFrame({'Quote': [text], 'Author': [author], 'Tags': [tags]})], ignore_index=True)

df
```

```python
final = pd.DataFrame(columns=['Quote', 'Author First Name', 'Author Last Name', 'Tags'])

for j in range(1, 11):
    webpage = requests.get(f'https://quotes.toscrape.com/page/{j}/').text
    soup = BeautifulSoup(webpage, 'lxml')

    quotes = soup.find_all('div', class_='quote')

    for quote in quotes:
        text = quote.find('span', class_='text').text
        authorFirst = quote.find('small', class_='author').text.split()[0]
        authorLast = quote.find('small', class_='author').text.split()[-1]
        tags = [tag.text for tag in quote.find_all('a', class_='tag')]
        final = pd.concat([final, pd.DataFrame({'Quote': [text], 'Author First Name': [authorFirst], 'Author Last Name': [authorLast], 'Tags': [tags]})], ignore_index=True)

final
```
